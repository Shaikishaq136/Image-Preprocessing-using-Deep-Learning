resizing the image pillow library

```python
from PIL import Image
```

```python
img = Image.open('/content/NationalGeographic_2572187_square.jpg')
```

```python
img_resized = img.resize((300, 300))
```

```python
img_resized.save('cat_image_resized.jpg')
```

```python


img_res = mpimg.imread('/content/cat_image_resized.jpg')
img_plot = plt.imshow(img_res)
plt.show()
```

Convrting RGB images to Grayscale image using OpenCV

```python
import cv2
```

```python
img = cv2.imread('/content/NationalGeographic_2572187_square.jpg')
type(img)
```

```python
img.shape
```

```python
grayscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

```python
grayscale_image.shape
```

```python
from google.colab.patches import cv2_imshow
```

```python
cv2_imshow(grayscale_image)
```

```python
#Save the grayscale image
```

```python
cv2.imwrite('cat_grayscale_image.jpg', grayscale_image)
```
